### Sentence and word tokenization using spacy (object oriented)

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp("Mr. Batman loves Bangalore dosa. superman loves jhansi's narayan chat")

for sentence in doc.sents:
    print(sentence)
```

```python
for sentence in doc.sents:
    for word in sentence:
        print(word)
```

### Sentence and word tokenization using NLTK (String Processing library)
- NLTK gives more options to choose manually will algorithm we want

```python
from nltk.tokenize import sent_tokenize
```

```python
sent_tokenize("Mr. Batman loves Bangalore dosa. superman loves jhansi's narayan chat")
```

```python
from nltk.tokenize import word_tokenize
word_tokenize("Mr. Batman loves Bangalore dosa. superman loves jhansi's narayan chat")
```
